```python
transaction_upd.groupby(['name', 'minute'], as_index=False)\
                .agg({'transaction': 'count'})\
                .rename(columns={'transaction':'number_transaction'})\
                .pivot(index='minute', columns='name', values='number_transaction')\
                .fillna(0)\
                .head()
```

```python
transaction_upd.pivot_table(columns='name', 
                            index='minute',
                            values='transaction',
                            aggfunc='count',
                            fill_value = 0)\
                  .head()
```
