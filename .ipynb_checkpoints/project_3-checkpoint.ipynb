```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
users_df = pd.read_csv('https://stepik.org/media/attachments/lesson/360348/user_data.csv')
users_df.head()
```

```python
logs_df = pd.read_csv('https://stepik.org/media/attachments/lesson/360348/logs.csv')
logs_df.head()
```

```python
logs_df.describe()
```

```python
users_df.describe()
```

```python
logs_df.platform.nunique()
```

```python
count_client = logs_df.query('success == True')\
                .groupby('client', as_index=False)\
                .agg({'success': 'count'})\
                .sort_values('success', ascending = False)
```

```python
successfull_clients = count_client[count_client.success == count_client.success.max()].client.tolist()
successfull_clients.sort()
print(', '.join([str(i) for i in successfull_clients]))
```

```python
count_df_2 = logs_df\
                .groupby('client', as_index=False)\
                .agg({'success': 'sum'})\
                .sort_values('success', ascending = False)\
                .rename(columns={'success': 'success_number'})
count_df_2 
```

```python
count_platform = logs_df\
                .groupby('platform', as_index=False)\
                .agg({'success': 'sum'})\
                .sort_values('success', ascending = False)\
                .rename(columns={'success': 'success_number'})
count_platform
```

```python
data = users_df.merge(logs_df) # merge(logs_df, how='left', on ='client')
data.query('premium == True')\
      .platform\
      .value_counts()\
      .idxmax()
```

```python
sns.distplot(data.query('premium == False').age)
sns.distplot(data.query('premium == True').age)
```

```python
sns.histplot(data.query('premium == False').age)
sns.histplot(data.query('premium == True').age)
```

```python
client_vs_succsess_number = data.groupby('client')\
                            .agg({'success': 'sum'})
```

```python
sns.distplot(client_vs_succsess_number, kde=False)
```
