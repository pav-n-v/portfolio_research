```python
import pandas as pd
bookings = pd.read_csv('bookings.csv', sep=';')
```

```python
def standart_columns(column):
  return column.lower().replace(' ', '_')
```

```python
bookings_head = bookings[:7]
bookings_head
```

```python
bookings.shape[1]
```

```python
bookings.dtypes#.value_counts()
```

```python
bookings = bookings.rename(columns=standart_columns)
```

```python
bookings.query('is_canceled == 0')\
        .country\
        .value_counts()[:5]
```

```python
bookings.groupby('hotel')\
        .agg({'stays_total_nights':'mean'})\
        .round(2)
```

```python
bookings.query('assigned_room_type != reserved_room_type or is_canceled == 1').shape[0]
```

```python
print(bookings.query('arrival_date_year == 2016')\
        .arrival_date_month\
        .value_counts())
print(bookings.query('arrival_date_year == 2017')\
        .arrival_date_month\
        .value_counts())
```

```python
bookings.query("is_canceled == 1 and hotel == 'City Hotel'")\
        .groupby('arrival_date_year')\
        ['arrival_date_month']\
        .value_counts()
```

```python
bookings[['adults','children','babies']].mean()
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
bookings.groupby('hotel')\
        .agg({'total_kids': 'mean'})\
        .round(2)
```

```python
bookings.query('is_canceled == 1 and has_kids == False').shape[0] / bookings.query('has_kids == False').shape[0]
bookings.query('is_canceled == 1 and has_kids == True').shape[0] / bookings.query('has_kids == True').shape[0]
```
