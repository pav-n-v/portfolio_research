```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
def read_n_agg(path):
  df = pd.read_csv(path, sep = ';')
  return df.groupby('company').agg({'income': 'mean'})
```

```python
taxi = pd.read_csv('https://stepik.org/media/attachments/lesson/359240/taxi_peru.csv', sep = ';', parse_dates = ['start_at', 'end_at', 'arrived_at'])
taxi
```

```python
taxi.source.value_counts(normalize = True).mul(100).round()[0]
```

```python
9741+7631+4909+571+136
```

```python
df_source = taxi.groupby('source', as_index = False)\
                .agg({'taxi_id': 'count'})\
                .rename(columns={'taxi_id': 'order_count'}).sort_values('order_count', ascending = False)
df_source
```

```python
sns.barplot(data = df_source, x = 'source', y = 'order_count')
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(taxi['source'], hue = taxi['end_state'])

```

```python
sns.countplot(taxi['icon'])
```

```python
driver_score_counts = taxi\
                      .driver_score\
                      .value_counts(normalize = True)\
                      .mul(100)\
                      .round(2)\
                      .reset_index()\
                      .rename(columns = {'index':'driver_score',
                                          'driver_score': 'percentage'})\
                      .sort_values('driver_score')
                
driver_score_counts
```

```python
ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage')
sns.despine()  # убрать часть рамки графика
plt.show()
```

```python
rider_score_counts = taxi\
                      .rider_score\
                      .value_counts(normalize = True)\
                      .mul(100)\
                      .round(2)\
                      .reset_index()\
                      .rename(columns = {'index':'rider_score',
                                          'rider_score': 'percentage'})\
                      .sort_values('rider_score')
                
rider_score_counts
```

```python
ax = sns.barplot(data=rider_score_counts, x='rider_score', y='percentage', color='red', alpha=0.5)
ax.set(xlabel='Rider score', ylabel='Percentage')
sns.despine()
```
